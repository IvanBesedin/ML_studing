```python

def plot_numerical_with_missing(df, charts_per_figure=12, bins=30):
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    
    if not num_cols:
        print("Числовые колонки не найдены.")
        return

    print(f"Найдено числовых колонок: {len(num_cols)}. Начинаем построение...\n")

    chunks = [num_cols[i:i + charts_per_figure] for i in range(0, len(num_cols), charts_per_figure)]
    
    sns.set_theme(style="whitegrid", context="notebook")

    for chunk_idx, cols_chunk in enumerate(chunks):
        n_plots = len(cols_chunk)
        n_cols = 4 if n_plots >= 4 else n_plots
        n_rows = (n_plots + n_cols - 1) // n_cols
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 5))
        if n_plots == 1:
            axes = [axes]
        else:
            axes = axes.flatten()
            
        total_rows = len(df)

        for i, col in enumerate(cols_chunk):
            try:
                missing_count = df[col].isna().sum()
                missing_percent = (missing_count / total_rows) * 100
                clean_data = df[col].dropna()

                color = "skyblue" if missing_percent < 50 else "salmon"

                axes[i].hist(clean_data, bins=bins, color=color, edgecolor='black', alpha=0.7)
                
                axes[i].set_ylabel("Частота (Count)", fontsize=10)
                axes[i].set_xlabel("Значение", fontsize=10)
                axes[i].grid(True, axis='y', linestyle='--', alpha=0.6)
                axes[i].tick_params(labelsize=8)
                
                title_text = f"{col}\nПропуски: {missing_percent:.2f}% ({missing_count:,})"
                axes[i].set_title(title_text, fontsize=11, pad=10, fontweight='bold')
                
                if len(clean_data) == 0:
                    axes[i].text(0.5, 0.5, "Нет данных для отображения", 
                                 transform=axes[i].transAxes, ha='center', va='center', color='red')
                
            except Exception as e:
                axes[i].text(0.5, 0.5, f"Ошибка:\n{str(e)}", transform=axes[i].transAxes, 
                             ha='center', va='center', color='red', fontsize=9, bbox=dict(facecolor='white', alpha=0.8))
                axes[i].set_title(f"{col} (Error)", fontsize=10, color='red')
                axes[i].axis('off')

        for j in range(n_plots, len(axes)):
            fig.delaxes(axes[j])
            
        plt.tight_layout(pad=2.0)
        plt.show()
        print(f"--- Отображена фигура {chunk_idx + 1} из {len(chunks)} ---\n")

plot_numerical_with_missing(data, charts_per_figure=12, bins=50)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_recall_curve, roc_auc_score, average_precision_score

results = {}

def evaluate_model(name, model, X_tr, X_te, y_tr, y_te, scaled=False):
    """Обучить, предсказать, сохранить результаты."""
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)
    
    # --- Получаем scores для ROC-AUC и PR ---
    if hasattr(model, "predict_proba"):
        # Для KNN, RandomForest, LogisticRegression
        y_scores = model.predict_proba(X_te)[:, 1]
    elif hasattr(model, "decision_function"):
        # Для SVM (LinearSVC, SVC)
        y_scores = model.decision_function(X_te)
    else:
        # Fallback (не должно случиться)
        y_scores = y_pred.astype(float)
    
    # --- Метрики ---
    acc = accuracy_score(y_te, y_pred)
    f1 = f1_score(y_te, y_pred, average='macro')
    
    # ROC-AUC (правильно — на scores)
    auc = roc_auc_score(y_te, y_scores)
    
    # PR-кривая и PR-AUC (правильно — на scores)
    precision, recall, thresholds = precision_recall_curve(y_te, y_scores)
    pr_auc = average_precision_score(y_te, y_scores)  # или auc(recall, precision)
    
    # --- Сохраняем всё ---
    results[name] = {
        'accuracy': acc, 
        'f1_macro': f1,
        'roc_auc': auc,
        'pr_auc': pr_auc,
        'precision': precision,  # массивы для кривой
        'recall': recall,
        'thresholds': thresholds
    }
    
    print(f'{name}: accuracy={acc:.4f}, F1-macro={f1:.4f}, ROC-AUC={auc:.4f}, PR-AUC={pr_auc:.4f}')
    
    # Дополнительно: confusion matrix для анализа
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(y_te, y_pred)
    print(f'Confusion matrix:\n{cm}\n')
    
    return y_pred
```

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
import time

BALANCED = {0: 1, 1: 10}

# SVM на подвыборке (полный датасет слишком большой для RBF/poly SVM)
np.random.seed(42)
idx_sub = np.random.choice(len(X_train), size=10000, replace=False)
X_train_sub = X_train[idx_sub]
y_train_sub = y_train.iloc[idx_sub].reset_index(drop=True)

# Нормализация данных (критично для SVM!)
scaler = StandardScaler()
X_train_sub_scaled = scaler.fit_transform(X_train_sub)
X_test_scaled = scaler.transform(X_test)

# Функция оценки с замером времени и подробной информацией
def evaluate_svm(name, model, X_tr, X_te, y_tr, y_te):
    print(f"\n{'='*50}")
    print(f"Обучение {name}...")
    start_time = time.time()
    
    model.fit(X_tr, y_tr)
    train_time = time.time() - start_time
    
    start_time = time.time()
    y_pred = model.predict(X_te)
    predict_time = time.time() - start_time
    
    accuracy = accuracy_score(y_te, y_pred)
    
    print(f"✓ {name} завершён")
    print(f"  Время обучения: {train_time:.2f} сек")
    print(f"  Время предсказания: {predict_time:.2f} сек")
    print(f"  Accuracy: {accuracy:.4f}")
    print(f"\nClassification Report:")
    print(classification_report(y_te, y_pred, digits=4))
    
    return y_pred, accuracy

results = {}

# 1. Линейный SVM (быстрый, через LinearSVC)
print("\n" + "="*50)
print("ТЕСТИРОВАНИЕ ВСЕХ ЯДЕР SVM")
print("="*50)

linear_svm = LinearSVC(
    C=1.0, 
    max_iter=10000,
    class_weight=BALANCED, 
    random_state=RANDOM_STATE,
    dual='auto'
)
y_pred_lsvm, acc_lsvm = evaluate_svm(
    'LinearSVM (LinearSVC)', 
    linear_svm, 
    X_train_sub_scaled, X_test_scaled, 
    y_train_sub, y_test
)
results['LinearSVC'] = acc_lsvm

# 2. Линейное ядро через SVC (для сравнения)
linear_svc_kernel = SVC(
    kernel='linear',
    C=1.0,
    class_weight=BALANCED,
    random_state=RANDOM_STATE
)
y_pred_linear, acc_linear = evaluate_svm(
    'Linear Kernel (SVC)',
    linear_svc_kernel,
    X_train_sub_scaled, X_test_scaled,
    y_train_sub, y_test
)
results['Linear (SVC)'] = acc_linear

# 3. RBF ядро (радиальная базисная функция)
rbf_svm = SVC(
    kernel='rbf',
    C=10,
    gamma='scale',
    class_weight= BALANCED,
    random_state=RANDOM_STATE
)
y_pred_rbf, acc_rbf = evaluate_svm(
    'RBF Kernel',
    rbf_svm,
    X_train_sub_scaled, X_test_scaled,
    y_train_sub, y_test
)
results['RBF'] = acc_rbf

# 4. Полиномиальное ядро (degree=3 по умолчанию)
poly_svm = SVC(
    kernel='poly',
    degree=3,
    C=10,
    gamma='scale',
    class_weight=BALANCED,
    random_state=RANDOM_STATE
)
y_pred_poly, acc_poly = evaluate_svm(
    'Polynomial Kernel (degree=3)',
    poly_svm,
    X_train_sub_scaled, X_test_scaled,
    y_train_sub, y_test
)
results['Poly (d=3)'] = acc_poly

# 5. Полиномиальное ядро с degree=2 (для сравнения)
poly2_svm = SVC(
    kernel='poly',
    degree=2,
    C=10,
    gamma='scale',
    class_weight=BALANCED,
    random_state=RANDOM_STATE
)
y_pred_poly2, acc_poly2 = evaluate_svm(
    'Polynomial Kernel (degree=2)',
    poly2_svm,
    X_train_sub_scaled, X_test_scaled,
    y_train_sub, y_test
)
results['Poly (d=2)'] = acc_poly2

# 6. Сигмоидное ядро (аналог нейронной сети)
sigmoid_svm = SVC(
    kernel='sigmoid',
    C=10,
    gamma='scale',
    class_weight=BALANCED,
    random_state=RANDOM_STATE
)
y_pred_sigmoid, acc_sigmoid = evaluate_svm(
    'Sigmoid Kernel',
    sigmoid_svm,
    X_train_sub_scaled, X_test_scaled,
    y_train_sub, y_test
)
results['Sigmoid'] = acc_sigmoid

# 7. Пользовательское ядро (пример: хи-квадрат ядро для гистограмм)
# Для изображений MNIST/CIFAR полезно косинусное сходство
def cosine_kernel(X, Y):
    """Косинусное сходство: K(X,Y) = <X,Y> / (||X|| * ||Y||)"""
    X_norm = np.linalg.norm(X, axis=1).reshape(-1, 1)
    Y_norm = np.linalg.norm(Y, axis=1).reshape(1, -1)
    return np.dot(X, Y.T) / (X_norm * Y_norm + 1e-10)



# Предварительно нормализуем для косинусного ядра (уже сделано через StandardScaler)
# Но для косинуса лучше L2-нормализация
X_train_l2 = X_train_sub_scaled / (np.linalg.norm(X_train_sub_scaled, axis=1, keepdims=True) + 1e-10)
X_test_l2 = X_test_scaled / (np.linalg.norm(X_test_scaled, axis=1, keepdims=True) + 1e-10)

custom_svm = SVC(
    kernel='precomputed',
    C=10,
    class_weight=BALANCED,
    random_state=RANDOM_STATE
)

# Вычисляем матрицы ядер (может занять время!)
print("\nВычисление косинусной матрицы ядра...")
K_train = cosine_kernel(X_train_l2, X_train_l2)
K_test = cosine_kernel(X_test_l2, X_train_l2)

y_pred_custom, acc_custom = evaluate_svm(
    'Custom Cosine Kernel (precomputed)',
    custom_svm,
    K_train, K_test,
    y_train_sub, y_test
)
results['Cosine (custom)'] = acc_custom

# Итоговое сравнение
print("\n" + "="*50)
print("СРАВНИТЕЛЬНАЯ ТАБЛИЦА РЕЗУЛЬТАТОВ")
print("="*50)
sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)
for i, (name, acc) in enumerate(sorted_results, 1):
    marker = "🏆" if i == 1 else "  "
    print(f"{marker} {i}. {name:<25} | Accuracy: {acc:.4f}")

print(f"\nЛучшее ядро: {sorted_results[0][0]} с accuracy {sorted_results[0][1]:.4f}")
```

```python
# ============================================
# 1. BAGGING (для лучшей модели - LGBM)
# ============================================
print("\n" + "=" * 70)
print("1. BAGGING (Bootstrap Aggregating)")
print("=" * 70)

# Bagging для LightGBM
bagging_lgbm = BaggingClassifier(
    estimator=lgbm,
    n_estimators=10,
    max_samples=0.95,
    max_features=0.8,
    bootstrap=True,
    bootstrap_features=False,
    n_jobs=-1,
    random_state=42
)

bagging_lgbm.fit(X_train, y_train)
y_pred_bag = bagging_lgbm.predict(X_test)
y_proba_bag = bagging_lgbm.predict_proba(X_test)[:, 1]

print("\nBagging (LGBM base):")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_bag):.4f}")
print(f"F1-macro:  {f1_score(y_test, y_pred_bag, average='macro'):.4f}")
print(f"ROC-AUC:   {roc_auc_score(y_test, y_proba_bag):.4f}")
print(f"PR-AUC:    {average_precision_score(y_test, y_proba_bag):.4f}")
```

```python
# ============================================
# 3. BLENDING (Advanced — оптимальный порог + разнообразие моделей)
# ============================================
print("\n" + "=" * 70)
print("3. BLENDING (Advanced — threshold tuning + diverse models)")
print("=" * 70)

from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve

# Разделение
X_blend_train, X_blend_val, y_blend_train, y_blend_val = train_test_split(
    X_train, y_train, test_size=0.3, random_state=42, stratify=y_train
)

# Добавляем разнообразных моделей (меньше корреляция!)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

base_models_diverse = [
    ('lgbm', lgbm),  # градиентный бустинг
    ('rf', rf),      # бэггинг деревьев
    ('lr', LogisticRegression(max_iter=1000, class_weight='balanced', n_jobs=-1)),  # линейная
    ('gnb', gnb)     # байесовская
]

meta_train = np.zeros((len(X_blend_val), len(base_models_diverse)))
meta_test = np.zeros((len(X_test), len(base_models_diverse)))

for i, (name, model) in enumerate(base_models_diverse):
    print(f"\n  [{i+1}/{len(base_models_diverse)}] {name.upper()}...")
    
    model.fit(X_blend_train, y_blend_train)
    
    if hasattr(model, "predict_proba"):
        meta_train[:, i] = model.predict_proba(X_blend_val)[:, 1]
        meta_test[:, i] = model.predict_proba(X_test)[:, 1]
    else:
        meta_train[:, i] = model.decision_function(X_blend_val) if hasattr(model, "decision_function") else model.predict(X_blend_val)
        meta_test[:, i] = model.decision_function(X_test) if hasattr(model, "decision_function") else model.predict(X_test)
    
    val_auc = roc_auc_score(y_blend_val, meta_train[:, i])
    print(f'      → val_roc_auc={val_auc:.4f}')

# Мета-модель с L2 регуляризацией (L1 убивает GNB, пробуем L2)
meta_clf = LogisticRegression(
    C=1.0,              # слабее регуляризация
    penalty='l2',
    class_weight='balanced',
    random_state=42,
    max_iter=1000
)
meta_clf.fit(meta_train, y_blend_val)

y_proba_blend = meta_clf.predict_proba(meta_test)[:, 1]

# ===== КЛЮЧЕВОЕ: подбор оптимального порога по F1 =====
precisions, recalls, thresholds = precision_recall_curve(y_test, y_proba_blend)
f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-8)
best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx] if best_idx < len(thresholds) else 0.5

print(f"\nОптимальный порог: {best_threshold:.3f} (F1={f1_scores[best_idx]:.4f})")

y_pred_blend = (y_proba_blend >= best_threshold).astype(int)

print(f"\n{'='*50}")
print("BLENDING ADVANCED RESULTS:")
print(f"{'='*50}")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_blend):.4f}")
print(f"F1-macro:  {f1_score(y_test, y_pred_blend, average='macro'):.4f}")
print(f"F1-class1: {f1_score(y_test, y_pred_blend, pos_label=1):.4f}")
print(f"ROC-AUC:   {roc_auc_score(y_test, y_proba_blend):.4f}")
print(f"PR-AUC:    {average_precision_score(y_test, y_proba_blend):.4f}")

print(f"\nВеса моделей:")
for i, (name, _) in enumerate(base_models_diverse):
    print(f"  {name}: {meta_clf.coef_[0][i]:.4f}")

# Корреляция
meta_df = pd.DataFrame(meta_train, columns=[n for n,_ in base_models_diverse])
print(f"\nКорреляция мета-признаков:")
print(meta_df.corr().round(3))
```
