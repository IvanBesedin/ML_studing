При решении задания имейте в виду, что под критерием Джини подразумевается такая функция:

$$Q(R) = -\frac {|R_{\ell}|}{|R|}H(R_\ell) -\frac {|R_r|}{|R|}H(R_r) ,$$

где $R$ — множество объектов, попавших в вершину, $R_{\ell}$ и $R_r$ — объекты, попавшие в левое и правое поддеревья,
$H(R) = 1 - p_1^2 - p_0^2$, $p_1$, $p_0$ — доли объектов класса 1 и 0 соответственно.

Для категориальных признаков применяется наивный алгоритм разбиения: мы пытаемся найти одно значение, разбиение по которому сильнее всего увеличит критерий информативности. Иными словами, объекты с конкретным значением признака отправляем в левое поддерево, остальные - в правое.

```python
def find_best_split(
    feature_vector: Union[np.ndarray, pd.DataFrame], 
    target_vector: Union[np.ndarray, pd.Series],
    task: str = "classification",
    feature_type: str = "real"
) -> Tuple[np.ndarray, np.ndarray, float, float]:
    """
    Указания:
    * Пороги, приводящие к попаданию в одно из поддеревьев пустого множества объектов, не рассматриваются.
    * В качестве порогов, нужно брать среднее двух соседних (при сортировке) значений признака
    * Поведение функции в случае константного признака может быть любым.
    * При одинаковых приростах Джини или дисперсии нужно выбирать минимальный сплит.

    :param feature_vector: вещественнозначный вектор значений признака
    :param target_vector: вектор классов объектов,  len(feature_vector) == len(target_vector)
    :param task: либо `classification`, либо `regression`
    :param feature_type: либо `real`, либо `categorical`
    
    :return thresholds: отсортированный по возрастанию вектор со всеми возможными порогами, по которым объекты можно
     разделить на две различные подвыборки, или поддерева
    :return ginis/variances: вектор со значениями критерия Джини для каждого из порогов в thresholds len(ginis) == len(thresholds)
    :return threshold_best: оптимальный порог (число)
    :return gini_best/variance_best: оптимальное значение критерия Джини/дисперсии (число)
    """
    
    if task == 'regression':
        
        
        
        return (thresholds, variances, threshold_best, variance_best)
    
    
    
    elif task == 'classification': #рассматриваем бинарную классификацию
        
        
    
        return (thresholds, ginis, threshold_best, gini_best)
               

    
```

**Выведите график зависимости значения критерия ошибки от порогового значения при разбиении вершины по признаку `CRIM`.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston

boston = load_boston()
X = pd.DataFrame(data=boston["data"], columns=boston["feature_names"])
y = boston["target"]
# X.describe()
X.head()
```

```python
unique_values = np.unique(X["RAD"])
unique_values
```

```python
def gini_group(y):
    if len(y) == 0:
        return 0
    p0 = sum(y == 0) / len(y)
    p1 = sum(y == 1) / len(y)
    H = 1 - p0**2 - p1**2
    return H


def gini_split(y_left, y_right, y_all):

    n_left = len(y_left)
    n_right = len(y_right)
    n_all = len(y_all)
    
    if n_left == 0 or n_right == 0:
        return -np.inf
    
    H_left = gini_group(y_left)
    H_right = gini_group(y_right)
    
    Q = -(n_left / n_all) * H_left - (n_right / n_all) * H_right
    
    return Q


def variance_split(y_left, y_right, y_all):

    n_left = len(y_left)
    n_right = len(y_right)
    n_all = len(y_all)
    
    if n_left == 0 or n_right == 0:
        return np.inf 
    
    var_left = np.var(y_left)
    var_right = np.var(y_right)
    
    weighted_var = (n_left * var_left + n_right * var_right) / n_all
    
    return weighted_var


def find_best_split(feature_vector, target_vector, task="classification", feature_type="real"):

    feature_vector = np.array(feature_vector).flatten()
    target_vector = np.array(target_vector).flatten()
    
    n = len(feature_vector)
    thresholds = []  
    scores = []      
    
    # ========================================================
    # Случай 1: Вещественный признак (real)
    # ========================================================
    if feature_type == "real":
        
        sorted_features = np.sort(feature_vector)
        unique_values = np.unique(sorted_features)
        
        if len(unique_values) <= 1:
            return (np.array([]), np.array([]), np.inf, np.inf)
        
        for i in range(len(unique_values) - 1):
            threshold = (unique_values[i] + unique_values[i + 1]) / 2
            thresholds.append(threshold)
        
        for threshold in thresholds:
            left_mask = feature_vector <= threshold
            right_mask = feature_vector > threshold
            
            y_left = target_vector[left_mask]
            y_right = target_vector[right_mask]
            
            if task == "regression":
                score = variance_split(y_left, y_right, target_vector)
            else:  # classification
                score = gini_split(y_left, y_right, target_vector)
            
            scores.append(score)
    
    # ========================================================
    # Случай 2: Категориальный признак (categorical)
    # ========================================================
    elif feature_type == "categorical":
        
        unique_values = np.unique(feature_vector)
        
        if len(unique_values) <= 1:
            return (np.array([]), np.array([]), np.inf, np.inf)
        
        for threshold in unique_values:
            thresholds.append(threshold)
            
            left_mask = feature_vector == threshold
            right_mask = feature_vector != threshold
            
            y_left = target_vector[left_mask]
            y_right = target_vector[right_mask]
            
            if task == "regression":
                score = variance_split(y_left, y_right, target_vector)
            else:
                score = gini_split(y_left, y_right, target_vector)
            
            scores.append(score)
    
    thresholds = np.array(thresholds)
    scores = np.array(scores)
    
    if len(scores) == 0:
        return (thresholds, scores, np.inf, np.inf)
    
    if task == "classification":
        best_idx = np.argmax(scores)  # Максимум (из-за минуса в формуле)
    else:
        best_idx = np.argmin(scores)  # Минимум дисперсии
    
    threshold_best = thresholds[best_idx]
    score_best = scores[best_idx]
    
    return (thresholds, scores, threshold_best, score_best)



# ============================================================
# Тест 1: Регрессия
# ============================================================

print("=" * 50)
print("Тест 1: Регрессия (признак CRIM)")
print("=" * 50)


thresholds, variances, threshold_best, variance_best = find_best_split(
    X["CRIM"].to_numpy(), 
    y, 
    task="regression",
    feature_type="real"
)

print(f"Найдено порогов: {len(thresholds)}")
print(f"Лучший порог: {threshold_best:.4f}")
print(f"Лучшая дисперсия: {variance_best:.4f}")

plt.figure(figsize=(10, 5))
plt.plot(thresholds, variances, label='Дисперсия', linewidth=2)
plt.axvline(x=threshold_best, color='red', linestyle='--', linewidth=2, 
            label=f'Лучший порог: {threshold_best:.2f}')
plt.xlabel('Порог')
plt.ylabel('Дисперсия')
plt.title('Зависимость дисперсии от порога (CRIM)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


# ============================================================
# Тест 2: Классификация
# ============================================================
print("\n" + "=" * 50)
print("Тест 2: Классификация (признак CRIM)")
print("=" * 50)

# Создаём бинарную целевую переменную
y_binary = (y > np.median(y)).astype(int)

thresholds_cls, ginis, threshold_best_cls, gini_best_cls = find_best_split(
    X["CRIM"].to_numpy(), 
    y_binary, 
    task="classification",
    feature_type="real"
)

print(f"Найдено порогов: {len(thresholds_cls)}")
print(f"Лучший порог: {threshold_best_cls:.4f}")
print(f"Лучший Q(R): {gini_best_cls:.4f}")



plt.figure(figsize=(10, 5))
plt.plot(thresholds_cls, ginis, label='Q(R)', color='green', linewidth=2)
plt.axvline(x=threshold_best_cls, color='red', linestyle='--', linewidth=2, 
            label=f'Лучший порог: {threshold_best_cls:.2f}')
plt.xlabel('Порог')
plt.ylabel('Q(R)')
plt.title('Зависимость Q(R) от порога (CRIM)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


print("\n" + "=" * 50)
print("✅ Всё работает!")
print("=" * 50)

# ============================================================
# Тест 3: Регрессия (категориальный признак RAD)
# ============================================================
print("\n" + "=" * 50)
print("Тест 3: Регрессия (категориальный признак RAD)")
print("=" * 50)

# Для категориального признака пороги — это сами категории
thresholds_rad_reg, variances_rad, threshold_best_rad_reg, variance_best_rad = find_best_split(
    X["RAD"].to_numpy(), 
    y, 
    task="regression",
    feature_type="categorical"
)

print(f"Найдено категорий (порогов): {len(thresholds_rad_reg)}")
print(f"Лучшая категория (порог): {threshold_best_rad_reg}")
print(f"Лучшая взвешенная дисперсия: {variance_best_rad:.4f}")

# Построение графика: для каждой категории — взвешенная дисперсия
plt.figure(figsize=(12, 5))
plt.bar(range(len(thresholds_rad_reg)), variances_rad, tick_label=thresholds_rad_reg)
plt.axhline(y=variance_best_rad, color='red', linestyle='--', linewidth=2,
            label=f'Лучшее значение: {variance_best_rad:.4f}')
plt.xlabel('Категория (RAD)')
plt.ylabel('Взвешенная дисперсия')
plt.title('Зависимость взвешенной дисперсии от категории признака RAD (регрессия)')
plt.legend()
plt.grid(True, axis='y', alpha=0.3)
plt.show()


# ============================================================
# Тест 4: Классификация (категориальный признак RAD)
# ============================================================
print("\n" + "=" * 50)
print("Тест 4: Классификация (категориальный признак RAD)")
print("=" * 50)

thresholds_rad_cls, ginis_rad, threshold_best_rad_cls, gini_best_rad = find_best_split(
    X["RAD"].to_numpy(), 
    y_binary, 
    task="classification",
    feature_type="categorical"
)

print(f"Найдено категорий (порогов): {len(thresholds_rad_cls)}")
print(f"Лучшая категория (порог): {threshold_best_rad_cls}")
print(f"Лучший Q(R): {gini_best_rad:.4f}")

plt.figure(figsize=(12, 5))
plt.bar(range(len(thresholds_rad_cls)), ginis_rad, tick_label=thresholds_rad_cls)
plt.axhline(y=gini_best_rad, color='red', linestyle='--', linewidth=2,
            label=f'Лучшее Q(R): {gini_best_rad:.4f}')
plt.xlabel('Категория (RAD)')
plt.ylabel('Q(R)')
plt.title('Зависимость Q(R) от категории признака RAD (классификация)')
plt.legend()
plt.grid(True, axis='y', alpha=0.3)
plt.show()

print("\n" + "=" * 50)
print("✅ Все тесты выполнены!")
print("=" * 50)
```
